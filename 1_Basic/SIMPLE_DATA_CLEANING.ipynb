```python
class BaseSalary:
    def __init__(self, base_salary = 20000, bonus_rate = 0.1, symbol = '$'):
        self.base_salary = base_salary
        self.bonus_rate = bonus_rate
        self.symbol = symbol
        
    def __repr__(obj):
        return f"{obj.symbol}{obj.base_salary:,.0f}"
    
    def calculate_salary(self):
        self.total_salary = self.base_salary * (1 + self.bonus_rate)
        return f"{self.symbol}{self.total_salary:,.0f}"
```

```python
# Import the libraries

import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Load the Dataset from HuggingFace
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Counts of Jobs

```python
df.pivot_table(index = 'job_title_short', aggfunc = 'size')
```

```python
df.pivot_table(index = 'job_title_short', aggfunc = 'count')
```

```python
df.value_counts('job_title_short')
```

```python
df.groupby('job_title_short').size()
```

# Median Salary by Country

```python
df.pivot_table(index = 'job_title_short', columns = 'job_country', values = 'salary_year_avg', aggfunc = 'median')
```

```python
df.pivot_table(index = 'job_country', columns = 'job_title_short', values = 'salary_year_avg', aggfunc = 'median')
```

```python
top_countries = df['job_country'].value_counts().head(6).index

top_countries
```

```python
# Create a pivot table, making the countries the index

df_job_country_salary = df.pivot_table(index = 'job_country', columns = 'job_title_short', values = 'salary_year_avg', 
                                       aggfunc = 'median')

df_job_country_salary = df_job_country_salary.loc[top_countries]
```

```python
df_job_country_salary
```

```python
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']

df_job_country_salary = df_job_country_salary[job_titles]

df_job_country_salary
```

```python
df_job_country_salary.plot(kind = 'bar')
plt.ylabel("Median Salary ($USD)")
plt.xlabel(" ")
plt.title("Median Salary by Country and Job Title")
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
df_job_country_salary2 = df.pivot_table(index = 'job_country', columns = 'job_title_short', values = 'salary_year_avg', 
                                       aggfunc = ['median', 'mean'])
```

```python
df_copy = df.copy()

df_jobtitle_country_size = df_copy.pivot_table(index = 'job_country', columns = 'job_title_short', 
                                                aggfunc = 'size', fill_value = 0)

df_jobtitle_country_size
```

```python
top_countries = df['job_country'].value_counts().head(6).index

top_countries
```

```python
df_jt_cs = df_jobtitle_country_size.loc[top_countries]

df_jt_cs
```

```python
data_jobs = ['Data Analyst', 'Data Engineer', 'Data Scientist']

df_jt_cs = df_jt_cs[data_jobs]

df_jt_cs
```

```python
df_jt_cs.plot(kind = 'bar')
plt.ylabel("Count of Jobs")
plt.xlabel(" ")
plt.title("Most Popular Data Jobs by Country and Job Title")
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

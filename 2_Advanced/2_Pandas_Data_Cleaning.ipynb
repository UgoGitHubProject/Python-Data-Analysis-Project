```python
# Import libraries

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Load Dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Clean Dataset
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Make a copy of the dataframe

df_filled = df.copy()   # Make a copy of the dataset using the .copy() function
df_original = df.copy() # Make another copy of the dataset using the .copy() function
df_original_1 = df.copy() # Make another copy of the dataset using the .copy() function
df_filled_2 = df   # Make a copy of dataset using assignment operation
```

```python
median_salary_year = df_filled['salary_year_avg'].median()  # Calculate the median of the salary_year column
median_salary_hour = df_filled['salary_hour_avg'].median()  # Calculate the median of the salary_hour column
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_salary_year) # Replace NaN values with median
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_salary_hour) # Replace NaN values with median
```

```python
df_unique = df_filled.copy()
df_unique2 = df_filled_2

df_unique.drop_duplicates(inplace = True) # Removing duplicate records inplace within the copy of dataframe
print('Length of original df:       ', len(df_filled))
print('Length of drop duplicates df:', len(df_unique))
print('Rows Dropped:'                , len(df_filled) - len(df_unique))
print(' ')

df_unique = df_unique.drop_duplicates()  # Removing duplicate records; assign back to the copy of dataframe
print('Length of original df:       ', len(df_filled))
print('Length of drop duplicates df:', len(df_unique))
print('Rows Dropped:'                , len(df_filled) - len(df_unique))
print(' ')



df_unique2.drop_duplicates(inplace = True)
print('Length of original df:       ', len(df_filled_2))
print('Length of drop duplicates df:', len(df_unique2))
print('Rows Dropped:'                , len(df_filled_2) - len(df_unique2))
print(' ')

df_unique2 = df_unique2.drop_duplicates()
print('Length of original df:       ', len(df_filled_2))
print('Length of drop duplicates df:', len(df_unique2))
print('Rows Dropped:'                , len(df_filled_2) - len(df_unique2))




```

```python
# Import the libraries

import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Load the Dataset from HuggingFace
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_USA_Software_pivot = pd.read_csv("https://lukeb.co/software_csv" , index_col = 'job_posted_month')

df_USA_Software_pivot
```

```python
df_USA = df[df['job_country'] == 'United States'].copy()

df_USA['job_posted_month'] = df_USA['job_posted_date'].dt.strftime('%B')

df_USA_pivot = df_USA.pivot_table(index = 'job_posted_month', columns = 'job_title_short', aggfunc = 'size')

df_USA_pivot = df_USA_pivot.reset_index()

df_USA_pivot['month_number'] = pd.to_datetime(df_USA_pivot['job_posted_month'], format = '%B').dt.month

df_USA_pivot.sort_values('month_number', inplace = True)

df_USA_pivot.set_index('job_posted_month' , inplace = True)

df_USA_pivot.drop(columns = 'month_number', inplace = True)

df_USA_pivot
```

```python
# Merge the two datasets; one of the Data Jobs the other of the Software Jobs

df_USA_merged = df_USA_pivot.merge(df_USA_Software_pivot, on = 'job_posted_month')

df_USA_merged
```

```python
top_5 = (df_USA_merged
         .sum()
         .sort_values(ascending = False)
         .head()
         .index
         .tolist()
)

top_5
```

```python
df_USA_merged[top_5].plot(kind = 'line')
plt.title("Job Postings of the Top Tech Jobs in USA")
plt.xlabel("2023")
plt.ylabel('Job Count')
plt.ylim(0, 22500)
plt.legend() ;
```
